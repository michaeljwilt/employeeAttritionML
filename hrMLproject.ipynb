```python
#Recoding Business Travel
def my_recode(BusinessTravel):
    if BusinessTravel == "Non-Travel":
        return 1
    if BusinessTravel == "Travel_Rarely":
        return 2
    if BusinessTravel == "Travel_Frequently":
        return 3


hr["BusinessTravel"].apply(my_recode)
```

```python
#Recoding Department
def my_recode2(Department):
    if Department == "Human Resources":
        return 1
    if Department == "Sales":
        return 2
    if Department == "Research & Development":
        return 3
hr["Department"].apply(my_recode2)
```

```python
hr1 = hr.drop("EducationField", axis = 1)
```

```python
hr2 = hr1.drop("StandardHours", axis = 1)
```

```python
hr2.head()
```

```python
#Data Wrangling
##Subset data into x and y for training model

x = hr2.drop("Attrition", axis = 1)
y = hr2["Attrition"]
```

```python
#Train Test Split
## Using 60/40 model for train/test
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.4, random_state=76)
```

```python
#Initial Random Forest Model
forest = RandomForestClassifier(n_estimators=500, random_state=76)
forest.fit(x_train, y_train)
```
